```python
def linear_cost(x, linear_cost_eq):
    return linear_cost_eq[0] * x + linear_cost_eq[1]
    
def quadratic_cost(x, quadratic_cost_eq):
    return quadratic_cost_eq[0] * (x**2) + quadratic_cost_eq[1] * x + quadratic_cost_eq[2]
```
